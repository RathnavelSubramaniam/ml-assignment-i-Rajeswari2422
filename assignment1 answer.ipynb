# b) What is the range of each quantitative predictor? You can answer this using the range() function. 

```python
col = ['mpg','cylinders','displacement','horsepower','weight','acceleration','year']
min = auto[col].min()
max = auto[col].max()
print(f"Range of {col}: {min} , {max}")
```

# C) What is the mean and standard deviation of each quantitative predictor? 

```python
mean = auto[col].mean()
std = auto[col].std()
print(f"{col}: {mean} , Std Dev of {col}: {std}")
```

# d) Now remove the 10th through 85th observations. What is the range, mean, and standard deviation of each predictor in the subset of the data that remains? 

```python
auto_subset = auto.drop(auto.index[9:85])
```

```python
min = auto_subset[col].min()
max = auto_subset[col].max()
mean = auto_subset[col].mean()
std = auto_subset[col].std()
print(f"Range of {col}: {min} , {max}")
print(f"Mean of {col}: {mean} , Std Dev of {col}: {std}")
```
